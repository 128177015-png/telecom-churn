```python
# outliers
```

```python
def cap_outliers(df, col):
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    bounds = [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR]
    df[col] = df[col].clip(lower = bounds[0], upper = bounds[1])
    print(IQR)

# Observations:
# 1. This function uses the Interquartile Range (IQR) method to define a "normal" range for your data.
# 2. Printing the IQR allows you to see the spread of the middle 50% of your data for that specific column.
```

```python
for col in df.iloc[:, 6:19]:
    cap_outliers(df, col)

# Observations:
# 1. This loop automates the cleaning process by applying the outlier function to 13 numerical columns at once.
# 2. removing rows, you preserve the size of your dataset.
# 3. Printing the IQR for each column helps you verify the natural spread of features.
```

```python
subset = df.iloc[:, 6:19]
n_cols = 3
n_rows = (subset.shape[1] + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize = (n_cols * 4, n_rows * 3), sharey = False)
axes = axes.flatten()

for i, col in enumerate(subset.columns):
    sns.histplot(subset[col], ax = axes[i], kde = True)
    axes[i].set_title(col)

for i in range(len(subset.columns), len(axes)):
    fig.delaxes(axes[i])

fig.suptitle("Distribution of Numerical Features")
plt.tight_layout()
plt.show()

# Observations:
# 1. The histograms now appear more focused and less stretched because the extreme outlier tails have been pulled inward.
# 2. You will notice taller bars at the very edges of some plots where multiple outlier values were "capped" to the same limit.
```

```python
df[['State', 'International plan', 'Voice mail plan']].head()

# Observations:
# 1. This selective preview focuses specifically on categorical features, showing how location and plan types are stored.
# 2. It reveals that 'International plan' and 'Voice mail plan' are stored as "Yes/No" strings, which may eventually need encoding into numbers.
```

```python
df["Churn"] = df["Churn"].astype("int64")
df["Churn"].dtype

# Observations:
# 1. This conversion changes the 'Churn' column from boolean True/False values to numerical 1s and 0s.
# 2. Converting to int64 is a necessary step for most machine learning models, as they require numerical inputs to perform calculations.
```

```python
d = {"No" : 0, "Yes" : 1}
df["Voice mail plan"] = df["Voice mail plan"].map(d)
df["International plan"] = df["International plan"].map(d)
df.head()

# Observations:
# 1. The map() function effectively transforms text labels into binary numbers, making the data readable for mathematical models.
# 2. This step completes the encoding for your "Yes/No" columns, ensuring all plan-related features are now in a consistent 0/1 format.
```

```python
# Univariate Analysis
```

```python
print("\nNumerical Summary\n", df.iloc[:, 6:21].describe())

# Observations:
# 1. This summary provides the statistical spread for all 15 usage-related columns, including your newly encoded binary features.
# 2. Since "International plan" and "Voice mail plan" are now 0s and 1s, the mean shows the percentage of users with each plan.
```

```python
categorical_cols = ['International plan', 'Voice mail plan']
for col in categorical_cols:
    print(f"\nFrequency for {col}:\n", df[col].value_counts() * 100)
    plt.figure(figsize = (8, 4))
    sns.countplot(data = df, x = col, palette = 'Set2')
    plt.title(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
    plt.tight_layout()

# Observations:
# 1. The printed percentages show that only a small minority of customers (around 9-10%) actually use the International plan.
# 2. The Voice mail plan is more common than the International plan but still used by less than a third of the total customer base.
```

```python
# Bivariate Analysis
```

```python
categorical_cols = ['International plan', 'Voice mail plan']
for col in categorical_cols:
    churn_by_col = df.groupby(col)["Churn"].mean().reset_index()
    print(f"\nChurn Rate by {col}\n", churn_by_col)
    sns.barplot(data = churn_by_col, x = col, y = 'Churn', palette = 'Set2')
    plt.title(f'Churn Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('Churn Rate')
    plt.show()
    plt.tight_layout()

# Observations:
# 1. Customers with an International Plan (1) have a significantly higher churn rate compared to those without it.
# 2. Conversely, the Voice Mail Plan seems to correlate with better retention, as those with the plan (1) show a lower churn average.
```
